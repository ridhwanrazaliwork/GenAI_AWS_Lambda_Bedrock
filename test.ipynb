```python
import json

# Example JSON string
json_string = '{"name": "Alice", "age": 30, "city": "New York"}'

# Using json.loads to parse JSON string to Python dict
data = json.loads(json_string)
print(data)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}
print(type(data))  # Output: <class 'dict'>

# Using json.dumps to serialize Python dict to JSON string
new_json = json.dumps(data)
print(new_json)  # Output: {"name": "Alice", "age": 30, "city": "New York"}
print(type(new_json))  # Output: <class 'str'>
```
